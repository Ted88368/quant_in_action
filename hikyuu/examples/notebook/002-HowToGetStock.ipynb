1 全局获取股票对象
==========

1.1 获取股票对象
-----------------

通过全局管理对象 sm，或使用函数 get_stock。股票标识格式“市场标识+股票代码”，市场标识：沪市sh，深市sz。

```python
#s = getStock('sh000001')
s = sm['sh000001']
print(s)
```

1.2 遍历所有股票
-----------------

```python
i = 0
# 遍历所有股票
for s in sm:
    i += 1
    #print(s)
print("全部数量:", i)

len(sm)
```

2 通过板块（Block）遍历股票对象
================

2.1 通过 sm.get_stock("板块分类", "板块名称") 获取相应板块
------------------------------------------------------------

```python
blk = sm.get_block("指数板块", "上证300")
for s in blk:
    if not s.valid:
        print(s)
```

2.1 获取自定义板块
------------------

自定义板块的板块分类固定为 “self”

```python
blk = sm.get_block("self", "1")
for s in blk:
    print(s)
```
