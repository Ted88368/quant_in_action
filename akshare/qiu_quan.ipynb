```python
import akshare as ak

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

#print(plt.style.available)
#plt.style.use('seaborn-v0_8-ticks')

index_fear_greed_funddb_df = ak.index_fear_greed_funddb()
print(index_fear_greed_funddb_df)

# 创建一个绘图对象
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_ylabel('上证指数', color='b')
ax1.plot(index_fear_greed_funddb_df['date'], index_fear_greed_funddb_df['index'], label = "上证指数")

# 创建第二个 Y 轴并绘制数据
ax2 = ax1.twinx()  # 使用 twinx() 共享 x 轴
ax2.set_ylabel('恐贪指数', color='b')
ax2.plot(index_fear_greed_funddb_df['date'], index_fear_greed_funddb_df['fear'],label ="恐贪指数", color = 'red')

# 添加图例
fig.tight_layout()  # 自动调整布局

plt.show()
```
