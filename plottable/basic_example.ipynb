# Basic Example

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from plottable import Table
```

```python
d = pd.DataFrame(np.random.random((5, 5)), columns=["A", "B", "C", "D", "E"]).round(2)

# figsize设置图表大小，单位为英寸
fig, ax = plt.subplots(figsize=(6, 3))

#使用dataframe创建表格
tab = Table(d)

plt.show()

fig.savefig("images/basic_table.png")
```

## Alternating Row Colors

```python
d = pd.DataFrame(np.random.random((5, 5)), columns=["A", "B", "C", "D", "E"]).round(2)

fig, ax = plt.subplots(figsize=(6, 3))

# row_dividers 是是否显示行分隔符，
# odd_row_color 是奇数行的颜色，even_row_color 是偶数行的颜色
tab = Table(d, row_dividers=False, odd_row_color="#f0f0f0", even_row_color="#e0f6ff")

plt.show()

fig.savefig("images/alternating_row_color.png")
```
