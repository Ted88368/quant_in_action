```python
import akshare as ak
import seaborn as sns
import matplotlib.pyplot as plt

# indicator="市盈率"; choice of {'市盈率', '市净率', '股息率', '风险溢价'}
index_value_hist_funddb_df = ak.index_value_hist_funddb(symbol="上证指数",indicator="市盈率")
df = index_value_hist_funddb_df.tail(60)
print(df)


# 使用 Seaborn 绘制图表，并指定 x 和 y 轴
sns.lineplot(x='日期', y='市盈率', data=df)

# 调整 X 轴文本距离和角度
plt.xticks(rotation=45)  # 旋转文本
plt.subplots_adjust(bottom=1)  # 调整底部间距

# 显示图表
plt.show()

```
