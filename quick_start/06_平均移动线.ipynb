```python
import baostock as bs
import pandas as pd
import mplfinance as mpf

# 登录baostock系统
lg = bs.login()

# 显示登陆返回信息
print('login respond error_code:'+lg.error_code)
print('login respond  error_msg:'+lg.error_msg)


# 设置股票代码和时间范围
stock_code = "sh.000001"  # 上证指数
start_date = "2023-06-01"
end_date = "2024-12-31"

# 获取股票数据
rs = bs.query_history_k_data_plus(stock_code,
                              "date,open,high,low,close,volume",
                              start_date=start_date,
                              end_date=end_date,
                              frequency="d",
                              adjustflag="3")  # adjustflag="3"表示不复权
data = rs.get_data()

# 登出baostock系统
bs.logout()

# 创建DataFrame
df = data[['date', 'open', 'high', 'low', 'close', 'volume']].rename(columns={'date': 'Date'})

# 转换日期格式
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)  # 设置日期列为索引
```

```python
# 将股价数据转换为数值类型
df[['open', 'high', 'low', 'close', 'volume']] = df[['open', 'high', 'low', 'close', 'volume']].astype(float)

df 

# 设置蜡烛图数据
ohlc = df[['open', 'high', 'low', 'close','volume']]
```

```python
# 计算多条简单移动平均线（SMA）数据
window1 = 50
window2 = 100
window3 = 200
ohlc['SMA1'] = ohlc['close'].rolling(window=window1).mean()
ohlc['SMA2'] = ohlc['close'].rolling(window=window2).mean()
ohlc['SMA3'] = ohlc['close'].rolling(window=window3).mean()
ohlc['SMA1']
```

```python

# 调用make_marketcolors函数，定义K线颜色
mc = mpf.make_marketcolors(
    up="red",  # 上涨K线的颜色
    down="green",  # 下跌K线的颜色
    edge="black",  # 蜡烛图箱体的颜色
    volume="blue",  # 成交量柱子的颜色
    wick="black"  # 蜡烛图影线的颜色
)
# 调用make_mpf_style函数，自定义图表样式
# 函数返回一个字典，查看字典包含的数据，按照需求和规范调整参数
style = mpf.make_mpf_style(base_mpl_style="ggplot", marketcolors=mc)
# print(style)



# 设置图形的大小
fig = mpf.figure()

# 绘制K线图和多条简单移动平均线
mpf.plot(ohlc, type='candle', style=style, volume=True, addplot=[
    mpf.make_addplot(ohlc['SMA1'], color='orange', label=f'SMA-{window1}', secondary_y=False),
    mpf.make_addplot(ohlc['SMA2'], color='blue', label=f'SMA-{window2}', secondary_y=False),
                     mpf.make_addplot(ohlc['SMA3'], color='red', label=f'SMA-{window3}', secondary_y=False)
], figratio=(10, 8))

# 显示图形
mpf.show()
```
